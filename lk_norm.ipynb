```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14, "figure.figsize": (6,5)})      
```

```python
def plotNormL(n, k):
    x = (2*(np.random.rand(2, n) - 0.5))
    y = (2*(np.random.rand(2, n) - 0.5))
    print(x)
    print(y)
    r = []
    for i in range(n):
        if np.linalg.norm(x[:,i]-y[:,i], ord=k) <=1:
            r.append(i)

    plt.fill(x[0, r]-y[0,r], x[1, r]-y[1, r], 'go')

    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.grid()
    plt.show()
```

```python
plotNormL(500,1)

#plotNormL(50,2)
#plotNormL(50,30)
#plotNormL(50,500)
```
